```python
X = full_df.select_dtypes(include='number')
y = full_df.select_dtypes('object') # defining y although 'songs' isn't my target
X
```

```python
X_pre = StandardScaler().fit(X)
X_prep = X_pre.transform(X)


pd.DataFrame(X_prep,columns=X.columns)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_prep)
```

```python
clusters = kmeans.predict(X_prep)

pd.Series(clusters).value_counts().sort_index()
```

```python
%%time
for i in range(len(X_prep[0])):
    plt.figure(figsize=(5, 5))
    plt.scatter(X_prep[:, 0], X_prep[:, i], c=clusters,alpha=0.5)
    plt.show()
```

```python
%%time
for i in range(len(X_prep[0])):
    plt.figure(figsize=(5, 5))
    plt.scatter(X_prep[:, 1], X_prep[:, i], c=clusters,alpha=0.5)
    plt.show()
```

```python
corr = pd.DataFrame(X_prep).corr()
plt.figure(figsize = (10,7))
sb.heatmap(corr, 
            cmap='RdBu_r',
            annot=False,
            linewidth=0.5)
plt.title('Correlation of in original data')
```

```python
%%time
kmeans = KMeans(n_clusters=3,
                init="random",
                n_init=100,  # try with 1, 4, 8, 20, 30, 100...
                max_iter=10,
                tol=0,
                random_state=42)
kmeans.fit(X_prep)
print(kmeans.inertia_)
```

```python
clusters = kmeans.predict(X_prep)

# Checking the size of the clusters. Small change to previous clusters
pd.Series(clusters).value_counts().sort_index()
```

### Finding the optimal cluster

```python
%%time
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=42)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')

```

```python
%%time
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=42)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')

# although there is a clear silhouette at 2, if I take into account the elbow method,
# with a second silhouette peak at 8, it would provider a more varied outcome. I will move forward with 8 clusters
```

```python
%%time
kmeans = KMeans(n_clusters=8,
                init="random",
                n_init=200, 
                max_iter=10,
                tol=0,
                random_state=42)
kmeans.fit(X_prep)
```

```python
clusters = kmeans.predict(X_prep)

pd.Series(clusters).value_counts().sort_index()
```

```python
#pd.DataFrame(clusters)
df_clustered = full_df
df_clustered['clusters'] = clusters
df_clustered['songart'] = (df_clustered.artist+' '+df_clustered.songs)
df_clustered
```

## The MVP

```python
# Turning off warning message https://www.delftstack.com/howto/python/suppress-warnings-python/
import warnings
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message')

# My top 100 hot songs
hot = pd.read_csv('..\\lab-web-scraping-single-page\\song_artist.csv')
hot.artist = hot.artist.str.title()
hot.song = hot.song.str.title()

df_clustered = pd.read_csv('..\\lab-unsupervised-learning-intro\\df_clustered.csv')
hot['songart'] = (hot.artist+' '+hot.song)

print("Do you want a recommendation for a 'song'? Please first enter the 'artist'")

artist = input()
artist = re.sub(r'\s+', ' ',artist).title().strip()

print('Great! You entered',artist,'Now enter a song')
song = input()
song = song.title().strip()
song_art = str(artist +' '+ song)

if song_art != ' ':
    
    if song_art in str(hot['songart']):
        hot2 =  hot[hot['song'] != song]
        random_i = random.randint(0,len(hot2['song'])-1)
        print('We recommond ',hot2['song'][random_i],'by ',hot2['artist'][random_i])
 
    else: # song_art not in str(hot['songart']):
        results = sp.search(q= song_art, limit=1)
        song_uri = results["tracks"]["items"][0]["uri"]
        spoti_features = sp.audio_features(song_uri)
        spoti_features = pd.DataFrame.from_dict(data = spoti_features, orient = 'columns')
        spoti_features_num = spoti_features.select_dtypes(include = np.number)
        spoti_features_array = spoti_features_num.to_numpy()
        spoti_features = X_pre.transform(spoti_features_array)
        song_prep_clus = kmeans.predict(spoti_features)
        random_index = random.randint(0, len(df_clustered['songs'][df_clustered['clusters']== song_prep_clus[0]])-1)
        random_index
        print('We recommond ', df_clustered['songs'][random_index], 'by', df_clustered['artist'][random_index])


else:
    print("no recommendation at this time. Please check you're input")
        
```
